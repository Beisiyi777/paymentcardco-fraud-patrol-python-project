```python
# Adapted functions, see references for sources #

# data_overview 
def data_overview(df):
    """
    Compile summary statistics for each column in a DataFrame, including data type, null values, non-null values, and unique values.

    Parameters
    ----------
    df : pd.DataFrame
        The input DataFrame for which column-level statistics will be compiled.

    Returns
    -------
    pd.DataFrame
        A DataFrame summarizing each column, including its name, data type, count of null values, non-null values, and number of unique values.
    
    Example
    -------
    >>> stats_df = data_overview(df)
    >>> print(stats_df)
    """
    summary_data = []
    
    for col in df.columns:
        col_dtype = df[col].dtype
        num_of_nulls = df[col].isnull().sum()
        num_of_non_nulls = df[col].notnull().sum()
        num_of_distinct_values = df[col].nunique()  
  
        summary_data.append({
            'column_name': col,
            'column_dtype': col_dtype,
            'number_of_nulls': num_of_nulls,
            'number_of_non_nulls': num_of_non_nulls,
            'number_of_distinct_values': num_of_distinct_values
        })

    summary_df = pd.DataFrame(summary_data)
    return summary_df

# iv_woe
def iv_woe(data, target, bins=10, show_woe=False):
    """
    Calculate Information Value (IV) and Weight of Evidence (WOE) scores for each independent variable with respect to a binary target variable.

    Important Notes:
    1) The Information Value (IV) score increases as the number of bins increases. Extra caution is advised when there are more than 20 bins, as some bins may have very few events or non-events.
    2) Information Value is typically used for binary logistic regression models, as it reflects conditional log odds.  It is not optimal for feature selection in models like Random Forests or Support Vector Machines (SVM).

    Parameters
    ----------
    data : pd.DataFrame
        The DataFrame containing both the independent variables (X) and the dependent variable (target).
    
    target : str
        Name of the binary target variable (1 = event, 0 = non-event).
    
    bins : int, optional
        Number of bins or intervals for numeric variables, default is 10.
    
    show_woe : bool, optional
        Whether to print the WOE calculation table, default is False.

    Returns
    -------
    newDF : pd.DataFrame
        A DataFrame containing the Information Value (IV) scores for each independent variable.
    
    woeDF : pd.DataFrame
        A DataFrame containing the Weight of Evidence (WOE) scores for each bin of each independent variable.
    
    Example
    -------
    >>> iv_df, woe_df = iv_woe(data=df, target='true_value', bins=10, show_woe=False)
    >>> print(iv_df, woe_df)
    """
    # Empty Dataframe
    newDF, woeDF = pd.DataFrame(), pd.DataFrame()
    
    # Extract Column Names
    cols = data.columns
    
    # Run WOE and IV on all the independent variables
    for ivars in cols[~cols.isin([target])]:
        # print("Processing variable:", ivars)
        # all numeric variables with unique values <= 10 are treated as categorical variables
        if (data[ivars].dtype.kind in 'bifc') and (len(np.unique(data[ivars]))>10):       
            binned_x = pd.qcut(data[ivars], bins,  duplicates='drop')
            d0 = pd.DataFrame({'x': binned_x, 'y': data[target]})
        else:
            d0 = pd.DataFrame({'x': data[ivars], 'y': data[target]})

        
        # Calculate the number of events in each group (bin)
        d = d0.groupby("x", as_index=False, observed=False).agg({"y": ["count", "sum"]})
        d.columns = ['Cutoff', 'N', 'Events']
        
        # Calculate % of events in each group
        d['% of Events'] = np.maximum(d['Events'], 0.5) / d['Events'].sum()

        # Calculate the non events in each group
        d['Non-Events'] = d['N'] - d['Events']

        # Calculate % of non events in each group
        d['% of Non-Events'] = np.maximum(d['Non-Events'], 0.5) / d['Non-Events'].sum()

        # Calculate WOE by taking natural log of division of % of non-events and % of events
        d['WoE'] = np.log(d['% of Events']/d['% of Non-Events'])
        d['IV'] = d['WoE'] * (d['% of Events'] - d['% of Non-Events'])
        d.insert(loc=0, column='Variable', value=ivars)
        # print("Information value of " + ivars + " is " + str(round(d['IV'].sum(),6)))
        temp =pd.DataFrame({"Variable" : [ivars], "IV" : [d['IV'].sum()]}, columns = ["Variable", "IV"])
        newDF=pd.concat([newDF,temp], axis=0)
        woeDF=pd.concat([woeDF,d], axis=0)

        # Show WOE Table
        if show_woe == True:
            print(d)
    
    return newDF, woeDF

# get_iv_color
def get_iv_color(iv):
    """
    Determine the color classification for the Information Value (IV) of a feature.

    Parameters
    ----------
    iv : float
        The Information Value (IV) score to evaluate.

    Returns
    -------
    str
        The color indicating the strength of the IV score:
        - 'gray' if iv < 0.02  (useless for prediction)
        - 'yellow' if 0.02 <= iv <= 0.1  (weak predictor)
        - 'orange' if 0.1 < iv <= 0.3  (medium predictor)
        - 'green' if 0.3 < iv <= 0.5  (strong predictor)
        - 'red' if iv > 0.5  (too good to be true)
    """
    if iv < 0.02:
        return 'gray'
    elif 0.02 <= iv <= 0.1:
        return 'yellow'
    elif 0.1 < iv <= 0.3:
        return 'orange'
    elif 0.3 < iv <= 0.5:
        return 'green'
    else:
        return 'red'

# feature_importance_sorted
def feature_importance_sorted(classification_model_input, X, y, model_name='model', feature_importances_input=None):
    """
    Calculate and sort feature importances for a given classification model.

    Parameters
    ----------
    classification_model_input : sklearn.base.ClassifierMixin or None
        A trained classification model to calculate feature importances from. If None, `feature_importances_input` must be provided.
    
    X : pd.DataFrame
        The feature matrix used for model fitting and importance calculation.
    
    y : pd.Series
        The target variable for model fitting.
    
    model_name : str, optional
        A name for the model, used to label the columns in the output DataFrame. Default is 'model'.
    
    feature_importances_input : array-like, optional
        Pre-calculated feature importances. If provided, the model will not be refitted.

    Returns
    -------
    pd.DataFrame
        A DataFrame with the feature importances and their corresponding ranks.
        Columns:
        - 'Feature': The name of the feature.
        - 'imp_{model_name}': The importance score of the feature.
        - 'rank_{model_name}': The rank of the feature based on its importance.
    """
    if classification_model_input is not None:
        model = classification_model_input
        model.fit(X, y)
        feature_importances = model.feature_importances_
    else:
        feature_importances = feature_importances_input

    feature_importances_sorted = sorted(zip (X.columns, feature_importances), key=lambda x: x[1], reverse=True)
    feature_importances_df = pd.DataFrame(feature_importances_sorted, columns=['Feature', f'imp_{model_name}'])

    feature_importances_df[f'rank_{model_name}'] = range(1, len(feature_importances_df) + 1)
    return feature_importances_df

# stratified_sampling
def stratified_sampling(data, strata_col, sample_size, ignore_index=False):
    """
    Create a random stratified sample from a DataFrame based on a specified column.

    Parameters
    ----------
    data : pd.DataFrame
        The input DataFrame from which the sample is drawn.
    
    strata_col : str
        The name of the column used to stratify the data.
    
    sample_size : int
        The total number of samples to be selected.
        
    ignore_index : bool, optional, default=True
        Flag to reset the index of the resulting sample or retain the original index from the input DataFrame.

    Returns
    -------
    pd.DataFrame
        A stratified sample of the input DataFrame.
    """
    groups = data.groupby(strata_col)
    sample_list = []
    sample_size_per_strata = np.round(sample_size / len(groups), 0).astype(int)

    for _, group in groups:
        stratum_sample = group.sample(n=sample_size_per_strata, replace=False, random_state=universal_seed)
        sample_list.append(stratum_sample)

    sample_df = pd.concat(sample_list, ignore_index=ignore_index)

    return sample_df
```

```python
# Original user-defined 'global' functions #

# test_for_normality
def test_for_normality(df, alpha=0.05):
    """
    Conduct a one-sample normality test on each numeric column in the given DataFrame.

    Parameters
    ----------
    df : pd.DataFrame
        The DataFrame containing at least one numeric column.
    
    alpha : float, optional
        The significance level for the test, default is 0.05.

    For DataFrames with more than 5000 rows, the Kolmogorov-Smirnov test is used.
    For DataFrames with 5000 rows or fewer, the Shapiro-Wilk test is used.

    Returns
    -------
    pd.DataFrame
        A DataFrame containing the column names, the test used, the p-value, the significance level, 
        and the result indicating whether each column is normally distributed.
    
    Notes
    -----
    The function assumes that all numeric columns in the DataFrame are continuous and that any missing values have already been handled.
    
    Example
    -------
    >>> test_for_normality(df, alpha=0.01)
    """
    normality_summary = []
    numeric_only = df.select_dtypes(include=[np.number]).columns

    for col in numeric_only:
        if len(df) > 5000:
            norm_test = 'Kolmogorov-Smirnov test'
            norm_test_pval = st.kstest(df[col], 'norm', args=(df[col].mean(), df[col].std()))[1]
        else:
            norm_test = 'Shapiro-Wilk test'
            norm_test_pval = st.shapiro(df[col])[1]
        
        if norm_test_pval > alpha:
            distribution_type = 'normally distributed'
        else:
            distribution_type = 'not normally distributed'
        
        normality_summary.append({
            'column': col,
            'test_used': norm_test,
            'p_value': norm_test_pval,
            'alpha': alpha,
            'distribution_type': distribution_type
        })

    normality_summary_df = pd.DataFrame(normality_summary)
    return normality_summary_df

# trained_model_metrics
def trained_model_metrics(trained_model, X, y, model_name=None, threshold=0.5):
    """
    Evaluate the performance of a trained classification model using a set of standard metrics, with an optional threshold for classifying label 1.

    This function assumes that the trained model has both 'predict' and 'predict_proba' methods.

    Parameters
    ----------
    trained_model : estimator object
        The trained machine learning model (e.g., scikit-learn model or pipeline) used to make predictions.
    
    X : array-like or pd.DataFrame, shape (n_samples, n_features)
        The feature matrix on which predictions will be made.
    
    y : array-like, shape (n_samples,)
        The true labels corresponding to the feature matrix X.
    
    model_name : str, optional
        An optional name for the model to include in the output summary, default is None. Useful when evaluating multiple models.
    
    threshold : float, optional
        The threshold for classifying a sample as the positive class (label = 1), default is 0.5. The threshold must be between 0 and 1.

    Returns
    -------
    pd.DataFrame
        A DataFrame containing the model name (if provided), and evaluation metrics including roc_auc, accuracy, precision, recall, false_positive_rate, and f1_score.
    
    Notes
    -----
    The function assumes that the classifier can produce probability estimates via 'predict_proba' and that it is a binary classifier.
    
    Example
    -------
    >>> metrics_df = trained_model_metrics(my_model, X_test, y_test, model_name="LogisticRegression")
    >>> print(metrics_df)
    """
    # predict and apply threshold
    y_pred_proba = trained_model.predict_proba(X)[:, 1]
    y_pred = (y_pred_proba >= threshold).astype(int)

    # generate confusion matrix
    tn, fp, fn, tp = skm.confusion_matrix(y, y_pred).ravel()
    fpr = fp / (tn + fp)

    # compile metrics
    metrics_summary = []
    metrics_summary.append({
        'model_name': model_name,
        'roc_auc': skm.roc_auc_score(y, y_pred_proba),
        'accuracy': skm.accuracy_score(y, y_pred),
        'precision': skm.precision_score(y, y_pred),
        'recall': skm.recall_score(y, y_pred),
        'false_positive_rate': fpr,
        'f_score': skm.f1_score(y, y_pred)
    })
    
    metrics_summary_df = pd.DataFrame(metrics_summary)
    return metrics_summary_df

# compile_metrics
def compile_metrics(y, y_pred_proba, y_pred, model_name=None):
    """
    Compile standard evaluation metrics from true labels, predicted probabilities, and predicted class labels.
    
    Parameters
    ----------
    y : array-like or pd.DataFrame, shape (n_samples,)
        True labels for the target variable.
    
    y_pred_proba : array-like or pd.DataFrame, shape (n_samples,)
        Predicted probabilities for the positive class (class 1).
    
    y_pred : array-like or pd.DataFrame, shape (n_samples,)
        Predicted class labels based on the applied threshold.
    
    model_name : str, optional
        Optional name for the model, included in the output summary. Useful when evaluating multiple models.

    Returns
    -------
    pd.DataFrame
        A DataFrame containing the compiled metrics, including model name (if provided), roc_auc, accuracy, precision, recall, false_positive_rate, and f_score.
    
    Notes
    -----
    - The function assumes binary classification with `predict_proba` available for computing ROC AUC.
    - `y_pred` should be derived from the predicted probabilities using an appropriate threshold.
    
    Example
    -------
    >>> metrics_df = compile_metrics(y_test, y_pred_proba, y_pred, model_name="RandomForest")
    >>> print(metrics_df)
    """
    # generate confusion matrix
    tn, fp, fn, tp = skm.confusion_matrix(y, y_pred).ravel()
    fpr = fp / (tn + fp)

    # compile metrics
    metrics_report = []
    metrics_report.append({
        'model_name': model_name,
        'roc_auc': skm.roc_auc_score(y, y_pred_proba),
        'accuracy': skm.accuracy_score(y, y_pred),
        'precision': skm.precision_score(y, y_pred),
        'recall': skm.recall_score(y, y_pred),
        'false_positive_rate': fpr,
        'f_score': skm.f1_score(y, y_pred)
    })
    
    metrics_report_df = pd.DataFrame(metrics_report)
    return metrics_report_df
```

```python
# Preprocessing #

# feature processing groups
onehot_features = ['tr_type', 'tr_state']  # for categorical features < 10 categories
target_features = ['tr_currency', 'user_country']  # for categorical features > 50 categories
scale_features = ['tr_amount_gbp', 'user_age', 'account_age', 'tr_day', 'tr_weekday', 'tr_hour']

# transformers
onehot_transformer = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False) 
target_transformer = TargetEncoder(target_type='binary', random_state=universal_seed)
scale_transformer = MinMaxScaler()

# preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', onehot_transformer, onehot_features),
        ('target', target_transformer, target_features),
        ('scale', scale_transformer, scale_features)
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
)

# fit and transform Train Set
X_train_preprocessed = preprocessor.fit_transform(X_train, y_train)

# convert preprocessed X_train to dataframe
X_train_preprocessed_df = pd.DataFrame(X_train_preprocessed, columns=preprocessor.get_feature_names_out(), index=X_train.index)

# convert preprocessed X_train discrete features to integers to conserve memory
discrete_features = [col for col in X_train_preprocessed_df.columns if any(cat in col for cat in onehot_features)] 
X_train_preprocessed_df[discrete_features] = X_train_preprocessed_df[discrete_features].apply(np.int32)

# consolidate preprocessed train set
train_set_preprocessed = pd.concat([X_train_preprocessed_df, y_train], axis=1)
```

```python
# Bayesian Optimization of xgboost, random forest, and support vector machine
# uses 16 of 16 features

# use half of train set to reduce computational load
X_train_a, X_train_b, y_train_a, y_train_b = train_test_split(X_train, y_train, test_size=0.5, stratify=y_train, random_state=universal_seed)

# use same feature processing groups as in Preprocessing step above ('onehot_features', 'target_features', 'scale_features')
# reinstantiate transformers and preprocessing steps for bayesian optimization to avoid any data leakage
onehot_transformer_bo = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False) 
target_transformer_bo = TargetEncoder(target_type='binary', random_state=universal_seed)
scale_transformer_bo = MinMaxScaler()

preprocessor_bo = ColumnTransformer(
    transformers=[
        ('onehot_bo', onehot_transformer_bo, onehot_features),
        ('target_bo', target_transformer_bo, target_features),
        ('scale_bo', scale_transformer_bo, scale_features)
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
)

# parameter functions for each model_type
def instantiate_xgb(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 400, step=100),
        'max_depth': trial.suggest_int('max_depth', 3, 7, step=2),
        'learning_rate': trial.suggest_categorical('learning_rate', [0.01, 0.1, 0.2, 0.3]),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 5, step=2),
        'random_state': universal_seed,
        'objective': 'binary:logistic'
    }
    return xgb.XGBClassifier(**params)

def instantiate_rf(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 400, step=100),
        'max_depth': trial.suggest_int('max_depth', 3, 7, step=2),
        'class_weight': trial.suggest_categorical('class_weight', [None, 'balanced']),
        'random_state': universal_seed
    }
    return RandomForestClassifier(**params)

def instantiate_svc(trial):
    params = {
        'C': trial.suggest_categorical('C', [100, 10, 1.0, 0.1, 0.01]),
        'kernel': trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly']),
        'gamma': trial.suggest_categorical('gamma', ['scale', 'auto']),
        'class_weight': trial.suggest_categorical('class_weight', [None, 'balanced']),
        'random_state': universal_seed
    }
    return SVC(**params)
```

```python
# objective function
def objective(trial):
    model_type = trial.suggest_categorical('model_type', ['xgb', 'rf', 'svc'])
    
    if model_type == 'xgb':
        model = instantiate_xgb(trial)
    elif model_type == 'rf':
        model = instantiate_rf(trial)
    elif model_type == 'svc':
        model = instantiate_svc(trial)
        
    # preprocess set and run model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor_bo),
        ('classifier', model)
    ])
    
    # use cross-validation
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=universal_seed)
    scores = cross_val_score(pipeline, X_train_a, y_train_a, cv=kfold, scoring='roc_auc')

    # report mean score for pruning review
    trial.report(np.mean(scores), step=0)

    # prune trial if poor performer
    if trial.should_prune():
        raise optuna.TrialPruned()
   
    return np.mean(scores)
```

```python
# create optuna study and optimize objective function
pruner = SuccessiveHalvingPruner(min_resource=1, reduction_factor=2, min_early_stopping_rate=0)
study = optuna.create_study(study_name='xgb-rf-svc-optimization', pruner=pruner, direction='maximize')
study.optimize(objective, n_trials=50, n_jobs=-1)

# output best trial
print("Best trial:")
best_trial = study.best_trial
print(f"  Value: {best_trial.value}")
print(f"  Params: {best_trial.params}")
```

```python
# Evaluate transaction-level performance at different thresholds

# calculate threshold candidates
# roc curve threshold (roc)
fpr, tpr, thresholds = skm.roc_curve(y_validation, y_pred_proba_best_model)
J_stat = tpr - fpr # Youden’s J statistic
ix_roc = np.argmax(J_stat)
threshold_roc = thresholds[ix_roc]

# precision-recall threshold (pr)
precision, recall, thresholds_pr = skm.precision_recall_curve(y_validation, y_pred_proba_best_model)
f_score = (2 * precision * recall) / (precision + recall)
ix_pr = np.argmax(f_score)
threshold_pr = thresholds_pr[ix_pr]

# TunedThresholdClassifierCV (ttcc)
ttcc = TunedThresholdClassifierCV(estimator=loaded_best_model, scoring='f1', cv=5)
ttcc.fit(X_validation, y_validation)
threshold_ttcc = ttcc.best_threshold_

# cost-sensitive threshold (cost)
P_train = train_set['is_fraudulent'].sum() # total number of actual positives in train set
N_train = len(train_set) - P_train # total number of actual negatives in train set

# assign costs with placeholders suggested by train set
fn_cost = np.round(train_set[train_set.is_fraudulent==1]['tr_amount_gbp'].max(),0) # stand-in loss value (max fraud value)
fp_cost = np.round(train_set[train_set.is_fraudulent==0]['tr_amount_gbp'].mean(),0) * 5 # stand-in customer friction value (5x mean non-fraud value)
tp_benefit = 5 # stand-in for reduced investigation time, other administrative overhead
tn_benefit = 10 # stand-in for not having to investigate, resolve customer concerns, other administrative overhead

# optimize cost threshold
costs = []
for p, r in zip(precision, recall):
    TP = r * P_train
    FP = TP / p - TP # derived from precision formula: Precision = TP / (TP + FP)
    FN = P_train - TP
    TN = N_train - FP

    cost = (fp_cost * FP) + (fn_cost * FN) - (tp_benefit * TP) - (tn_benefit * TN)
    costs.append(cost)

# identify threshold that minimizes cost
ix_cost = np.argmin(costs)
threshold_cost = thresholds_pr[ix_cost]

# calculate precision, recall at each threshold candidate
# roc
y_pred_roc = (y_pred_proba_best_model >= threshold_roc).astype(int)
precision_roc = skm.precision_score(y_validation, y_pred_roc)
recall_roc = skm.recall_score(y_validation, y_pred_roc)

# pr
precision_pr = precision[ix_pr]
recall_pr = recall[ix_pr]

# ttcc
y_pred_ttcc = (y_pred_proba_best_model >= threshold_ttcc).astype(int)
precision_ttcc = skm.precision_score(y_validation, y_pred_ttcc)
recall_ttcc = skm.recall_score(y_validation, y_pred_ttcc)

# cost
precision_cost = precision[ix_cost]
recall_cost = recall[ix_cost]

# calculate precision, recall at default threshold
ix_pr_default = np.argmax(thresholds_pr >= 0.5)
threshold_default = thresholds_pr[ix_pr_default]
precision_default = precision[ix_pr_default]
recall_default = recall[ix_pr_default]

# print results
print('Transaction-level thresholds\n')
print('Default Threshold=%f' % (threshold_default))
print('Best ROC Threshold=%f' % (threshold_roc))
print('Best Precision-Recall Threshold=%f' % (threshold_pr))
print('Best TunedThresholdClassifierCV Threshold=%f' % (threshold_ttcc))
print('Best Cost-Sensitive Threshold=%f' % (threshold_cost))
```

### <font color = 'purple'> Explain Patrol class with SHapley Additive exPlanations (SHAP) values

```python
# preprocess test_list
test_list_preprocessed = garrison.model.named_steps['preprocessor'].transform(test_list.drop(columns=['tr_id', 'user_id', 'is_fraudulent']))

# re-associate feature names, index, and transaction id
feature_names = garrison.model.named_steps['preprocessor'].get_feature_names_out() # extract feature names
test_list_preprocessed_df = pd.DataFrame(test_list_preprocessed, columns=feature_names, index=test_list.index) # convert to dataframe, retain original index
test_list_preprocessed_df['tr_id'] = test_list['tr_id']

# subset for feature columns only
feature_columns_test_list_preprocessed_df = test_list_preprocessed_df.drop(columns=['tr_id'])

# initialize shap explainer
explainer = shap.Explainer(garrison.model.named_steps['classifier'].predict, feature_columns_test_list_preprocessed_df)

# generate SHAP values
shap_values = explainer(feature_columns_test_list_preprocessed_df)

# convert shap values to dataframe
shap_values_df = pd.DataFrame(shap_values.values, index=test_list_preprocessed_df.index, columns=feature_names)
```
